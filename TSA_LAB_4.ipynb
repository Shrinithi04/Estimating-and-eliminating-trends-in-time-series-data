```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=15, freq='D')
temperature = np.linspace(25, 30, 15) + np.random.randn(15)
humidity = np.linspace(60, 65, 15) + np.random.randn(15)
pressure = np.linspace(1010, 1012, 15) + np.random.randn(15)
weather_data = pd.DataFrame({
    'Temperature': temperature,
    'Humidity': humidity,
    'Pressure': pressure
}, index=dates)
```

```python
rolling_mean = weather_data.rolling(window=3).mean()
trend_removed = weather_data - rolling_mean
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(weather_data.index, weather_data['Temperature'], label='Original Temperature', color='tab:blue', marker='o')
plt.scatter(weather_data.index, weather_data['Humidity'], label='Original Humidity', color='tab:orange', marker='x')
plt.scatter(weather_data.index, weather_data['Pressure'], label='Original Pressure', color='tab:green', marker='s')
plt.title('Original Weather Data (with Trend)')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
```

```python
plt.subplot(1, 2, 2)
plt.scatter(trend_removed.index, trend_removed['Temperature'], label='Processed Temperature', color='tab:blue', marker='o')
plt.scatter(trend_removed.index, trend_removed['Humidity'], label='Processed Humidity', color='tab:orange', marker='x')
plt.scatter(trend_removed.index, trend_removed['Pressure'], label='Processed Pressure', color='tab:green', marker='s')
plt.title('Processed Weather Data (Trend Removed)')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Original Data:\n", weather_data)
print("\nProcessed Data (Trend Eliminated):\n", trend_removed)
```
